```python
df = data.drop(["Row ID","Order ID","Customer ID","Product ID"],axis=1)
df.shape
```

```python
df[df["Postal Code"].isnull()]
```

```python
df["Postal Code"] = df["Postal Code"].fillna(5402)
df.isnull().sum()
```

```python
df[df["City"]=="Burlington"] 
```

```python
df["Months"] = df["Order Date"].str[3:5]
df
```

```python
df["Months"] = df["Months"].astype("int32")
df
```

```python
#ayları isimlere dönüştürme
d = {1:"Ocak",2:"Şubat",3:"Mart",4:"Nisan",5:"Mayıs",6:"Haziran",7:"Temmuz",8:"Ağustos",9:"Eylül",10:"Ekim",11:"Kasım",12:"Aralık"}
```

```python
d[4]
```

```python
df["Months"]=[d[i] for i in df["Months"]]
df
```

```python
#order date datetime yaptım
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")
df
```

```python
#ship date datetime dönüştürdüm
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")
df
```

```python
#tabloya yıl ekledik
df["Year"] = df["Order Date"].dt.year
df
```

```python
#tabloya gün ekledik
df["Day"] = df["Order Date"].dt.day
df
```
